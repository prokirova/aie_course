```python
import numpy as np
```

```python
# 2. Примеры создания массивов
list_data = [1, 2, 3, 4]
arr_from_list = np.array(list_data)
print(arr_from_list)
```

```python
nested_list = [[1, 2], [3, 4]]
arr_2d = np.array(nested_list)
print(arr_2d)
```

```python
zeros_arr = np.zeros(5)
print(zeros_arr)
```

```python
ones_arr = np.ones((2, 3))
print(ones_arr)
```

```python
arange_arr = np.arange(0, 10, 2)
print(arange_arr)
```

```python
print("Форма массива arr_2d:", arr_2d.shape)
print("Тип данных элементов:", arr_2d.dtype)
```

```python
# Индексы и срезы
element = arr_2d[0, 1]
print("Элемент [0,1]:", element)
```

```python
subarray = arr_2d[:, 1]
print("Срез [:,1] (второй столбец):", subarray)
```

```python
slice = arange_arr[1:3]
print("Срез arange_arr[1:3]:", slice)
```

```python
# Примеры векторных операций
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
```

```python
sum_arr = a + b
print("Поэлементное сложение a + b:", sum_arr)
```

```python
prod_arr = a * b
print("Поэлементное умножение a * b:", prod_arr)
```

```python
mean_val = np.mean(a)
print("Среднее значение массива a:", mean_val)
```

```python
total_sum = np.sum(b)
print("Сумма элементов массива b:", total_sum)
```
